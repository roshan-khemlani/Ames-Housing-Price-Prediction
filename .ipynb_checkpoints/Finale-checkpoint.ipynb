#### Features Engineering - Categorical Features (Nominal)
___

```python
# MS subclass is an int, convert to object
train['ms_subclass'] = train['ms_subclass'].astype(str)
test['ms_subclass'] = test['ms_subclass'].astype(str)
```

```python
# Create Nominal columns variable
# https://www.youtube.com/watch?v=Fm1F_dVMgEk&list=PLYGvOQTnahYKInRCq3_Q4viVR56dHX2x5&index=2
#nom_col = list(train.select_dtypes(include=['object']).columns)
nom_col = list(train.select_dtypes(include=['object']).columns)
```

```python
print(nom_col)
```

```python
# Check for 
# https://www.youtube.com/watch?v=Fm1F_dVMgEk&list=PLYGvOQTnahYKInRCq3_Q4viVR56dHX2x5&index=2
for col in nom_col:
    # impute mode for missing value in categorical columns
    train[col] = train[col].fillna(train[col].mode()[0])
    test[col] = test[col].fillna(test[col].mode()[0])
    
    train_values = sorted(list(train[col].unique()))
    test_values = sorted(list(test[col].unique()))
    
    # create unique set of categories 
    categories = set(train_values + test_values)
    
    train[col] = pd.Categorical(train[col], categories= categories)
    train[col] = pd.Categorical(train[col], categories= categories)
```

```python
# Create train and test dummy sets
train_2 = pd.get_dummies(train[nom_col])
test_2 = pd.get_dummies(test[nom_col])
```
