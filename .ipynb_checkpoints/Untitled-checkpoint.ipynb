___

```python
# Check for null values in train set
null_values = train.isnull().sum().sort_values(ascending=False)
print(null_values[null_values > 0])
```

```python
# Check for null values in train set
null_values2 = test.isnull().sum().sort_values(ascending=False)
print(null_values2[null_values2 > 0])
```

```python
# Drop id and pid columns as they are uniques identifiers, wont have any effect on saleprice
train.drop(['id', 'pid'], axis = 1, inplace = True)
test.drop(['id', 'pid'], axis = 1, inplace = True)
# Drop columns alley, pool_qc, fence and misc_feature 
# For both train and test data sets due to high number of null values (greater than 80%)
train.drop(['alley', 'pool_qc', 'fence', 'misc_feature'], axis = 1, inplace = True)
test.drop(['alley', 'pool_qc', 'fence', 'misc_feature'], axis = 1, inplace = True)
```

```python
# Fill null values of numerical column lot_frontage with column mean
train['lot_frontage'] = train['lot_frontage'].fillna(train['lot_frontage'].mean())
test['lot_frontage'] = test['lot_frontage'].fillna(test['lot_frontage'].mean())

# Fill null values of mas_vnr_type with mode as its catergorical
train['mas_vnr_type'] = train['mas_vnr_type'].fillna(train['mas_vnr_type'].mode()[0])
test['mas_vnr_type'] = test['mas_vnr_type'].fillna(test['mas_vnr_type'].mode()[0])

# Fill null values of numerical column mas_vnr_area with 0 as the mode category is None
train['mas_vnr_area'] = train['mas_vnr_area'].fillna(0)
test['mas_vnr_area'] = test['mas_vnr_area'].fillna(0)

# Fill null values of fireplace_qu with 0, as looking at the data as those properties dont have a fireplace 
train['fireplace_qu'] = train['fireplace_qu'].fillna(0)
test['fireplace_qu'] = test['fireplace_qu'].fillna(0)
```

- The null values for garage_yr_blt, garage_finish, garage_cond, garage_qual and garage_type is valid, as looking at the data  those properties dont have a garage hence fill with '0'.

```python
# Fill null values of column garage_yr_blt with 0, as looking at the data as those properties dont have a garage 
train['garage_yr_blt'] = train['garage_yr_blt'].fillna(0)
test['garage_yr_blt'] = test['garage_yr_blt'].fillna(0)

# Fill null values of column garage_yr_blt with 0, as looking at the data as those properties dont have a garage 
train['garage_finish'] = train['garage_finish'].fillna(0)
test['garage_finish'] = test['garage_finish'].fillna(0)

# Fill null values of column garage_cond with 0, as looking at the data as those properties dont have a garage 
train['garage_cond'] = train['garage_cond'].fillna(0)
test['garage_cond'] = test['garage_cond'].fillna(0)

# Fill null values of column garage_yr_blt with 0, as looking at the data as those properties dont have a garage 
train['garage_qual'] = train['garage_qual'].fillna(0)
test['garage_qual'] = test['garage_qual'].fillna(0)

# Fill null values of column garage_yr_blt with 0, as looking at the data as those properties dont have a garage 
train['garage_type'] = train['garage_type'].fillna(0)
test['garage_type'] = test['garage_type'].fillna(0)
```

- Notice for the ordinal columns Bsmt Qual, Bsmt Cond, Bsmt Exposure, BsmtFin Type 1 and BsmtFin Type 2 there are 54 rows with NA meaning those properties have no basement. Will fill the 

Notice for the ordinal columns Bsmt Qual, Bsmt Cond, Bsmt Exposure, BsmtFin Type 1 and BsmtFin Type 2 there are 54 rows with NA meaning those properties have no basement. Instead of filling in nan, will amend it later while switching the ordinal columns to numerical. 


```python
# Replace 'Duplex' with 'Duplx' 
train['bldg_type'] = train['bldg_type'].str.replace('Duplex','Duplx')
test['bldg_type'] = test['bldg_type'].str.replace('Duplex','Duplx')
# Check the amendments
print(np.unique(train['bldg_type']))
print(np.unique(test['bldg_type']))
```

```python
# Remove (agr) and (all) as per categories provided
train['ms_zoning'] = train['ms_zoning'].str.replace('A \(agr\)', 'A')
test['ms_zoning'] = test['ms_zoning'].str.replace('A \(agr\)', 'A')
train['ms_zoning'] = train['ms_zoning'].str.replace('C \(all\)', 'C')
test['ms_zoning'] = test['ms_zoning'].str.replace('C \(all\)', 'C')
train['ms_zoning'] = train['ms_zoning'].str.replace('I \(all\)', 'I')
test['ms_zoning'] = test['ms_zoning'].str.replace('I \(all\)', 'I')
# Check the amendments
print(np.unique(train['ms_zoning']))
print(np.unique(test['ms_zoning']))
```

```python
# Amend Change Wd Shng to WdShing and Change Brk Cmn to BrkComm
train['exterior_2nd'] = train['exterior_2nd'].str.replace('Wd Shng','WdShing')
test['exterior_2nd'] = test['exterior_2nd'].str.replace('Wd Shng','WdShing')
train['exterior_2nd'] = train['exterior_2nd'].str.replace('Brk Cmn','BrkComm')
test['exterior_2nd'] = test['exterior_2nd'].str.replace('Brk Cmn','BrkComm')
# Check the amendments
print(np.unique(train['exterior_2nd']))
print(np.unique(test['exterior_2nd']))
```

#### Data Cleaning - Collinear Features

```python
# Check the correlation matrix across the dataset for numerical features, drop features that are highly correlated
train_num = train._get_numeric_data()
corr_mat = train_num.corr().abs()
train_set = corr_mat.unstack()
train_set.sort_values(ascending = False).drop_duplicates()
```

```python
# Drop garage_cars column to avoid multi-collinearity, as it's highly correlated with garage_area
train.drop(['garage_cars'], axis = 1, inplace = True)
test.drop(['garage_cars'], axis = 1, inplace = True)
```

#### Data Cleaning - Combine Features and Interactive Terms

```python
# Combine bath rooms above ground full_bath and half_bath , rename ttl_bath_abv
train['ttl_bath_abv'] = train['full_bath'] + (0.5 * train['half_bath'])
test['ttl_bath_abv'] = test['full_bath'] + (0.5 * test['half_bath'])

# Combine basement bath rooms bsmt_full_bath and bsmt_half_bath, rename ttl_bath_base
train['ttl_bath_base'] = train['bsmt_full_bath'] + (0.5 * train['bsmt_half_bath'])
test['ttl_bath_base'] = test['bsmt_full_bath'] + (0.5 * test['bsmt_half_bath'])

# Combine yr_sold and yr_buit to see how old the property is, rename to prop_age
train['prop_age'] = train['yr_sold'] - train['year_built']
test['prop_age'] = test['yr_sold'] - test['year_built']
```

```python
# Check to see if sum of values in bsmtfin_sf_1,bsmtfin_sf_2 and bsmt_unf_sf equals to total_bsmt_sf
# If true drop bsmtfin_sf_1,bsmtfin_sf_2 and bsmt_unf_sf 
train['total'] = train['bsmtfin_sf_1'] + train['bsmtfin_sf_2'] + train['bsmt_unf_sf']
check = train['total'] == train['total_bsmt_sf']
check.value_counts()
```

#### Features Engineering - Convert Ordinal to Numerical
___

```python
# Change Ordinal values to numerical
# Column 'overall_qual' and 'overall_cond' are already int

# lot_shape
train['lot_shape'] = train['lot_shape'].map({'IR3': 1, 'IR2': 2, 'IR1': 3, 'Reg': 4})
test['lot_shape'] = test['lot_shape'].map({'IR3': 1, 'IR2': 2, 'IR1': 3, 'Reg': 4})
# utilities
train['utilities'] = train['utilities'].map({'ELO': 1, 'NoSeWa': 2, 'NoSewr': 3, 'AllPub': 4})
test['utilities'] = test['utilities'].map({'ELO': 1, 'NoSeWa': 2, 'NoSewr': 3, 'AllPub': 4})
# land_slope
train['land_slope'] = train['land_slope'].map({'Sev': 1, 'Mod': 2, 'Gtl': 3})
test['land_slope'] = test['land_slope'].map({'Sev': 1, 'Mod': 2, 'Gtl': 3})
# exter_qual
train['exter_qual'] = train['exter_qual'].map({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5 })
test['exter_qual'] = test['exter_qual'].map({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5 })
# exter_cond
train['exter_cond'] = train['exter_cond'].map({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5 })
test['exter_cond'] = test['exter_cond'].map({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5 })
# bsmt_qual
train['bsmt_qual'] = train['bsmt_qual'].map({ 'NA': 0,'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5 })
test['bsmt_qual'] = test['bsmt_qual'].map({ 'NA': 0,'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5 })
# bsmtf_cond
train['bsmt_cond'] = train['bsmt_cond'].map({ 'NA': 0,'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5 })
test['bsmt_cond'] = test['bsmt_cond'].map({ 'NA': 0,'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5 })
# bsmt_exposure
train['bsmt_exposure'] = train['bsmt_exposure'].map({ 'NA': 0,'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4 })
test['bsmt_exposure'] = test['bsmt_exposure'].map({'NA': 0,'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4})
# bsmtfin_type_1
train['bsmtfin_type_1'] = train['bsmtfin_type_1'].map({ 'NA': 0,'Unf': 1, 'LwQ': 2, 'Rec': 3, 'BLQ': 4, 'ALQ': 5, 'GLQ':6 })
test['bsmtfin_type_1'] = test['bsmtfin_type_1'].map({ 'NA': 0,'Unf': 1, 'LwQ': 2, 'Rec': 3, 'BLQ': 4, 'ALQ': 5, 'GLQ':6 })
# bsmtfin_type_2
train['bsmtfin_type_2'] = train['bsmtfin_type_2'].map({ 'NA': 0,'Unf': 1, 'LwQ': 2, 'Rec': 3, 'BLQ': 4, 'ALQ': 5, 'GLQ':6 })
test['bsmtfin_type_2'] = test['bsmtfin_type_2'].map({ 'NA': 0,'Unf': 1, 'LwQ': 2, 'Rec': 3, 'BLQ': 4, 'ALQ': 5, 'GLQ':6 })
# heating_qc
train['heating_qc'] = train['heating_qc'].map({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5 })
test['heating_qc'] = test['heating_qc'].map({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5 })
# electrical
train['electrical'] = train['electrical'].map({'Mix': 1, 'FuseP': 2, 'FuseF': 3, 'FuseA': 4, 'SBrkr': 5 })
test['electrical'] = test['electrical'].map({'Mix': 1, 'FuseP': 2, 'FuseF': 3, 'FuseA': 4, 'SBrkr': 5 })
# kitchen_qual
train['kitchen_qual'] = train['kitchen_qual'].map({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5 })
test['kitchen_qual'] = test['kitchen_qual'].map({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5 })
# functional
train['functional'] = train['functional'].map({'Sal': 1, 'Sev': 2, 'Maj2': 3, 'Maj1': 4, 'Mod': 5, 'Min2': 6, 'Min1': 7, 'Typ': 8 })
test['functional'] = test['functional'].map({'Sal': 1, 'Sev': 2, 'Maj2': 3, 'Maj1': 4, 'Mod': 5, 'Min2': 6, 'Min1': 7, 'Typ': 8 })
# garage_finish
train['garage_finish'] = train['garage_finish'].map({ 'NA': 0,'Unf': 1, 'RFn': 2, 'Fin': 3 })
test['garage_finish'] = test['garage_finish'].map({ 'NA': 0,'Unf': 1, 'RFn': 2, 'Fin': 3 })
# garage_qual
train['garage_qual'] = train['garage_qual'].map({ 'NA': 0,'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5 })
test['garage_qual'] = test['garage_qual'].map({ 'NA': 0,'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5 })
# garage_cond
train['garage_cond'] = train['garage_cond'].map({ 'NA': 0,'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5 })
test['garage_cond'] = test['garage_cond'].map({ 'NA': 0,'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5 })
# paved_drive
train['paved_drive'] = train['paved_drive'].map({ 'N': 1, 'P': 2, 'Y': 3 })
test['paved_drive'] = test['paved_drive'].map({ 'N': 1, 'P': 2, 'Y': 3 })
# fireplace quality
train['fireplace_qu'] = train['fireplace_qu'].map({ 'NA': 0,'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5 })
test['fireplace_qu'] = test['fireplace_qu'].map({ 'NA': 0,'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5 })
```

```python
def scatter_plot(dataframe, x_val, y_val, list_of_titles, list_of_xlabels, list_of_ylabels):
    # You'll want to specify your figsize
    fig, ax = plt.subplots(5, 4, figsize = (15,7))
    # Set titles, labels, etc here for each subplot
    ax.scatter(x=dataframe[x_val],y=dataframe[y_val], s = 50)
    ax.set_title(list_of_titles[0], fontsize = 12)
    ax.set_ylabel('saleprice', fontsize = 12)
    ax.set_xlabel(list_of_xlabels[0], fontsize = 12)
    ax.tick_params(axis="x", labelsize=12)
    ax.tick_params(axis="y", labelsize=12)
```
