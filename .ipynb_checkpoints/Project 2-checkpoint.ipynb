#### Data Cleaning - Removing Null Values
---

```python
# Drop pid columns as they are uniques identifiers, wont have any effect on saleprice
train.drop(['pid'], axis = 1, inplace = True)
test.drop(['pid'], axis = 1, inplace = True)
# Drop columns alley, pool_qc, fence and misc_feature 
# For both train and test data sets due to high number of null values (greater than 80%)
train.drop(['alley', 'pool_qc', 'fence', 'misc_feature'], axis = 1, inplace = True)
test.drop(['alley', 'pool_qc', 'fence', 'misc_feature'], axis = 1, inplace = True)
```

```python
# Fill null values of numerical column lot_frontage with column mean
train['lot_frontage'] = train['lot_frontage'].fillna(train['lot_frontage'].mean())
test['lot_frontage'] = test['lot_frontage'].fillna(test['lot_frontage'].mean())

# Fill null values of mas_vnr_type with mode as its catergorical
train['mas_vnr_type'] = train['mas_vnr_type'].fillna(train['mas_vnr_type'].mode()[0])
test['mas_vnr_type'] = test['mas_vnr_type'].fillna(test['mas_vnr_type'].mode()[0])

# Fill null values of numerical column mas_vnr_area with 0 as the mode category is None
train['mas_vnr_area'] = train['mas_vnr_area'].fillna(0)
test['mas_vnr_area'] = test['mas_vnr_area'].fillna(0)
```

- The null values for garage_yr_blt, garage_finish, garage_cond and garage_qual is valid, as looking at the data  those properties dont have a garage hence fill will '0' during Feature Engineering.
- garage_type is a nominal value will leave NA for now as its valid similar to above. 
- fireplace_quality null values is also valid, as those homes have no fireplace

- Notice for the ordinal columns Bsmt Qual, Bsmt Cond, Bsmt Exposure, BsmtFin Type 1 and BsmtFin Type 2 there are 54 rows with NA meaning those properties have no basement hence will replace with '0' during Feature Engineering as well.

```python
# Notice the data for row 1329 in train data is missing is value for all bsmt columns as it doesnt have a basement
# Notice for row 618 in train is missing the value for bsmt_full_bath and half_bath columns as it too doesnt have a basement
# Notice the data for row 1714 in train is missing is value for both garage_car and garage_area as it doesnt have a garage
# Will replace the remaining missing values with 0, take note during feature engineering for ordinal columns
# In test data, row 637 is missing the value in electrical column fill with mode since its categorical
test['electrical'] = test['electrical'].fillna(test['electrical'].mode()[0])
```

#### Data Cleaning - Update Incorrect Values in Columns

```python
# Replace 'Duplex' with 'Duplx' 
train['bldg_type'] = train['bldg_type'].str.replace('Duplex','Duplx')
test['bldg_type'] = test['bldg_type'].str.replace('Duplex','Duplx')
# Check the amendments
print(np.unique(train['bldg_type']))
print(np.unique(test['bldg_type']))
```

```python
# Remove (agr) and (all) as per categories provided
train['ms_zoning'] = train['ms_zoning'].str.replace('A \(agr\)', 'A')
test['ms_zoning'] = test['ms_zoning'].str.replace('A \(agr\)', 'A')
train['ms_zoning'] = train['ms_zoning'].str.replace('C \(all\)', 'C')
test['ms_zoning'] = test['ms_zoning'].str.replace('C \(all\)', 'C')
train['ms_zoning'] = train['ms_zoning'].str.replace('I \(all\)', 'I')
test['ms_zoning'] = test['ms_zoning'].str.replace('I \(all\)', 'I')
# Check the amendments
print(np.unique(train['ms_zoning']))
print(np.unique(test['ms_zoning']))
```

```python
# Amend Change Wd Shng to WdShing and Change Brk Cmn to BrkComm
train['exterior_2nd'] = train['exterior_2nd'].str.replace('Wd Shng','WdShing')
test['exterior_2nd'] = test['exterior_2nd'].str.replace('Wd Shng','WdShing')
train['exterior_2nd'] = train['exterior_2nd'].str.replace('Brk Cmn','BrkComm')
test['exterior_2nd'] = test['exterior_2nd'].str.replace('Brk Cmn','BrkComm')
# Check the amendments
print(np.unique(train['exterior_2nd']))
print(np.unique(test['exterior_2nd']))
```

### Feature Engineering
---

#### Combine Features and Interactive Terms
___

```python
# Combine bath rooms above ground full_bath and half_bath , rename ttl_bath_abv
train['ttl_bath_abv'] = train['full_bath'] + (0.5 * train['half_bath'])
test['ttl_bath_abv'] = test['full_bath'] + (0.5 * test['half_bath'])

# Combine basement bath rooms bsmt_full_bath and bsmt_half_bath, rename ttl_bath_base
train['ttl_bath_base'] = train['bsmt_full_bath'] + (0.5 * train['bsmt_half_bath'])
test['ttl_bath_base'] = test['bsmt_full_bath'] + (0.5 * test['bsmt_half_bath'])

# Combine yr_sold and yr_buit to see how old the property is, rename to prop_age
train['prop_age'] = train['yr_sold'] - train['year_built']
test['prop_age'] = test['yr_sold'] - test['year_built']
```

```python
# Check to see if sum of values in bsmtfin_sf_1,bsmtfin_sf_2 and bsmt_unf_sf equals to total_bsmt_sf
train['total'] = train['bsmtfin_sf_1'] + train['bsmtfin_sf_2'] + train['bsmt_unf_sf']
check = train['total'] == train['total_bsmt_sf']
check.value_counts()
```

```python
# If true drop bsmtfin_sf_1,bsmtfin_sf_2, bsmt_unf_sf and additonal column created 'total' for testing purpose in train set
train.drop(['bsmtfin_sf_1','bsmtfin_sf_2', 'bsmt_unf_sf', 'total'], axis = 1, inplace= True)
test.drop(['bsmtfin_sf_1','bsmtfin_sf_2', 'bsmt_unf_sf'], axis = 1, inplace= True)
```

```python
# Check to see if sum of values in 1st_flr_sf, 2nd_flr_sf and low_qual_fin_sf equals to gr_liv_area
train['living'] = train['1st_flr_sf'] + train['2nd_flr_sf'] + train['low_qual_fin_sf']
check_2 = train['living'] == train['gr_liv_area']
check_2.value_counts()
```

```python
# If true drop 1st_flr_sf, 2nd_flr_sf, low_qual_fin_sf and additonal column created 'living' for testing purpose in train set
train.drop(['1st_flr_sf','2nd_flr_sf', 'low_qual_fin_sf', 'living'], axis = 1, inplace= True)
test.drop(['1st_flr_sf','2nd_flr_sf', 'low_qual_fin_sf'], axis = 1, inplace= True)
```

```python
# Combine total bsmt sf and gr liv area for total floor size of property, rename to total_flr_siz
train['total_flr_size'] = train['total_bsmt_sf'] + train['gr_liv_area']
test['total_flr_size'] = test['total_bsmt_sf'] + test['gr_liv_area']
```

- Notice quite a few features with high collinearity in the above matrix, to avoid multicollinearity will drop those features.

#### Features Engineering - Categorical Features (Nominal)
___

```python
# MS subclass is an int, convert to object
train['ms_subclass'] = train['ms_subclass'].astype(str)
test['ms_subclass'] = test['ms_subclass'].astype(str)
# Replace '0' with NA for garage type as it was amended during data cleaning
train['garage_type'] = train['garage_type'].replace(0 , 'NA').astype(str)
test['garage_type'] = test['garage_type'].replace(0 , 'NA').astype(str)
```

```python
print(train.info())
print()
print(test.info())
```

```python
nom_col = list(train.select_dtypes(include=['object']).columns)
for col in nom_col:
    # impute mode for missing value in categorical columns
    train[col] = train[col].fillna(train[col].mode()[0])
    test[col] = test[col].fillna(test[col].mode()[0])
    # ensure the values are the same in both train and test set
    train_values = sorted(list(train[col].unique()))
    test_values = sorted(list(test[col].unique()))
    # create unique set of categories 
    categories = set(train_values + test_values)
    train[col] = pd.Categorical(train[col], categories= categories)
    train[col] = pd.Categorical(train[col], categories= categories)
```

```python
# Create Nominal column list for train and test
# nom_col_train = list(train.select_dtypes(include=['object']).columns)
# nom_col_test = list(test.select_dtypes(include=['object']).columns)
# Create train and test dummy sets
train_2 = pd.get_dummies(train[nom_col])
test_2 = pd.get_dummies(test[nom_col])
# Check to see the number of new columns including dummies
print(train_2.shape)
print(test_2.shape)
```

```python
# Check whats the missing values compared to both data sets
# Training dataset has additional 15 categories that are missing in Test, while all Test categories are present in Training
print(train_2.columns.difference(test_2.columns))
print(test_2.columns.difference(train_2.columns))
```

```python
# Combine the 2 sets to copy the columns from nominal train set to test
final_nom=pd.concat([train_2,test_2])
final_nom.head()
```

```python
# Check if the final data sets column matches train set 
train_2.columns.difference(final_nom.columns)
```

```python
# Now split the final data set back to the train and test data with the same columns
train_final = final_nom.iloc[:2049,:]
test_final = final_nom.iloc[2049:,:]
print(train_final.shape)
print(test_final.shape)
```

```python
# For the train/test final set we have only the nominal columns, retrieve the numerical columns as well 
numeric_train = train.select_dtypes(include=['float64','int64'])
numeric_test = test.select_dtypes(include=['float64','int64'])
# Combine to obtain the complete data set
df_train = pd.concat([train_final, numeric_train], axis=1, join='outer')
df_test = pd.concat([test_final, numeric_test], axis=1, join='outer')
# Check the shape of final data set for the both train and test
print(df_train.shape)
print(df_test.shape)
```

```python
# Check for null values in df_test set
null_values = df_test.isnull().sum().sort_values(ascending=False)
print(null_values[null_values > 0])
```

```python
# Replace null values with 0
df_test = df_test.replace(np.nan, 0)
```

#### Create the Train and Test Split Data

```python
# Input predictor variable
y = df_train['saleprice']
X = df_train.loc[:, df_train.columns != 'saleprice']

# Split the data into the training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Dimension of X_train, X_test
print(X_train.shape)
print(X_test.shape)
```

#### Feature Scaling

```python
# Standardizing the features by using Standard Scaler to fit X_train and transform X_train and X_test
ss = StandardScaler()
X_train_sc = ss.fit_transform(X_train)
X_test_sc = ss.transform(X_test)
submission = ss.fit_transform(df_test)
```

#### Finding Optimal Value

```python
lr = LinearRegression()
lr.fit(X_train_sc, y_train)
```

```python
# Finding Optimal value for Lasso
optimal_lasso = LassoCV(n_alphas=200, cv=10, verbose=1)
optimal_lasso.fit(X_train_sc, y_train)
print(optimal_lasso.alpha_)
```

```python
# Finding Optimal value for ridge
ridge_alphas = np.linspace(0.1, 10, 200)
optimal_ridge = RidgeCV(alphas=ridge_alphas, cv=10)
optimal_ridge.fit(X_train_sc, y_train)
print(optimal_ridge.alpha_)
```

```python
# Finding Optimal value for Elastic Net
l1_ratios = np.linspace(0.01, 1.0, 25)
optimal_enet = ElasticNetCV(l1_ratio=l1_ratios, n_alphas=200, cv=10,
                            verbose=1)
optimal_enet.fit(X_train_sc, y_train)
print(optimal_enet.alpha_)
print(optimal_enet.l1_ratio_)
```

#### Cross Validation

```python
# Linear Regression
# Cross validate R2 of linear  
lr_scores = cross_val_score(lr, X_train_sc, y_train, cv=10)
print(lr_scores)
print(np.mean(lr_scores))
```

```python
# Lasso
# Cross Validate Lasso R2 with the optimal alpha
lasso = Lasso(alpha=414)
lasso_scores = cross_val_score(lasso, X_train_sc, y_train, cv=10)
print(lasso_scores)
print(np.mean(lasso_scores))
```

```python
# Ridge
# Cross Validate Ridge R2 with the optimal alpha
ridge = Ridge(alpha=10)
ridge_scores = cross_val_score(ridge, X_train_sc, y_train, cv=10)
print(ridge_scores)
print(np.mean(ridge_scores))
```

```python
# ElasticNet
# Instantiate Elastic Net with optimal alpha
enet = ElasticNet(alpha=414, l1_ratio=1.0)
enet_scores = cross_val_score(enet, X_train_sc, y_train, cv=10)
print(enet_scores)
print(np.mean(enet_scores))
```
